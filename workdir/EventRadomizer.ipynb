```python
# Notebook to study the gammas emmited from the 2nubb events
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# Notebook to read in the nexus files and only save the useful stuff to slim the files down
import os
import sys
import tables as tb
import numpy  as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import re
from matplotlib.colors import LogNorm  # Import LogNorm for logarithmic color scale
from matplotlib.cm import get_cmap
```

```python
mbb_hits  = pd.read_hdf("/Users/mistryk2/Packages/nexus/mbb_test.h5", "MC/hits")
LQ_hits  = pd.read_hdf("/Users/mistryk2/Packages/nexus/Leptoquark_test.h5", "MC/hits")
```

```python


Nevents = 100


set_10_A = []
set_10_B = []
set_100_A = []
set_100_B = []
set_1000_A = []
set_1000_B = []

columns_to_keep = ["x","y","z","energy"]

for index, id in enumerate(mbb_hits.event_id.unique()):

    event_A = mbb_hits[mbb_hits.event_id == id]
    event_A = event_A[columns_to_keep]
    event_A["id"] = index

    event_B = LQ_hits[LQ_hits.event_id == id]
    event_B = event_B[columns_to_keep]
    event_B["id"] = index

    if (index < 10):
        set_10_A.append(event_A)
        set_10_B.append(event_B)

    if (index >=10 and index <110):
        set_100_A.append(event_A)
        set_100_B.append(event_B)

    if (index >=110 and index <1110):
        set_1000_A.append(event_A)
        set_1000_B.append(event_B)

    if (index >1110):
        break
    

set_10_A = pd.concat(set_10_A, ignore_index=True)
set_10_B = pd.concat(set_10_B, ignore_index=True)

set_100_A = pd.concat(set_100_A, ignore_index=True)
set_100_B = pd.concat(set_100_B, ignore_index=True)

set_1000_A = pd.concat(set_1000_A, ignore_index=True)
set_1000_B = pd.concat(set_1000_B, ignore_index=True)

def WriteFiles(df, filename):

    # Open the HDF5 file in write mode
    with pd.HDFStore(f"../files/{filename}.h5", mode='w', complevel=5, complib='zlib') as store:
        # Write each DataFrame to the file with a unique key
        store.put('MC/hits',df, format='table')

print("Writing Files...")
WriteFiles(set_10_A, "set_10_A")
WriteFiles(set_10_B, "set_10_B")

WriteFiles(set_100_A, "set_100_A")
WriteFiles(set_100_B, "set_100_B")

WriteFiles(set_1000_A, "set_1000_A")
WriteFiles(set_1000_B, "set_1000_B")
```
